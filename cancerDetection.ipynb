```python
# splitting the dataset into independent(X) and dependent(Y) dataset
X = df.iloc[:, 2:31].values
Y = df.iloc[:,1].values
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# splitting data into 25% testing and 75% trainging dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state = 5)
```

```python
Y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# scale the data (feature scaling)
sc =StandardScaler()
X_test= sc.fit_transform(X_test)
X_train= sc.fit_transform(X_train)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def scoreOftheModel(k):
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train, Y_train)
    return neigh.score(X_test,Y_test)    
```

```python
# creating a list of k with the score of the model
ls =[]
ls.append(("k","score"))
for i in range (1,50):
    ls.append((i,scoreOftheModel(i)))
ls
```

```python
# at k=7 model performs the best
```

```python
neigh = KNeighborsClassifier(n_neighbors=7)
```

```python
neigh.fit(X_train, Y_train)
```

```python
# checking score of the model on train dataset
neigh.score(X_train,Y_train)
```

```python
# testing model accuracy on test data on Confusion metriz for the model ([true positive],[false positive]
#                                                                        [false negative],[true negative])
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(Y_test, neigh.predict(X_test))
TP = cm[0][0]
TN = cm[1][1]
FN = cm[1][0]
FP = cm[0][1]
print(cm)
print("Testing Accuracy = ",(TP+TN)/(TP+TN+FP+FN))
```
